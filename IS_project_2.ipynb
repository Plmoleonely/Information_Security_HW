文字轉ascii函式

```python
def to_ascii(text):
  ascii_values = [ord(character) for character in text]
  return ascii_values
```

**project 3 文字位移4格函式**

```python
def ascii_plus4(text):
  for n in range(len(text)):
    while text[n] > 123 :
      text[n] += 4
      text[n] %= 128
    else :
      text[n] += 4
  return text
```

ascii轉二進位

```python
def to_bin(text):
  bin_values = []
  for n in range(len(text)):
    ans = ''
    list = []
    while text[n] > 0 :
      list.insert(0, text[n] % 2)
      text[n] //= 2
    for i in list:
      ans += str(i)
    bin_values.append(ans)
  return bin_values
```

二進位轉回ascii函式

```python
def b_to_a(text):
  a_values = []

  for n in range(len(text)):    
    counter = 1
    ans = 0
    text[n] = int(float(text[n]))
    while text[n] > 0:
      ans += (text[n] % 10) * counter
      text[n] //= 10
      counter *= 2

    a_values.append(ans)
  return a_values
```

ascii轉回文字函式

```python
def to_char(text):
  textchar = [chr(character) for character in text]
  return textchar
```

# 執行程式(文字)

```python
f = open('Txt_File .txt', 'r')
print(f)

text = f.read()
print(text)

ans_ascii = to_ascii(text)
print(ans_ascii)

ans_plus4 = ascii_plus4(ans_ascii)
print(ans_plus4)

ans_bin = to_bin(ans_plus4)
print(ans_bin)
```

```python
ans_b2a = b_to_a(ans_bin)
print(ans_b2a)
```

```python
ans_char = to_char(ans_b2a)
char = ''
for i in range(len(ans_char)):
  char += ans_char[i]
print(char)
```

**圖片平移四單位函式project_3**

```python
def img_plus4(img):
  for n in range(len(img)):
    while img[n] > 251 :
      img[n] += 4
      img[n] %= 256
    else :
      img[n] += 4
  return img
```
